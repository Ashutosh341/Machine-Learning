```python
#creating segmentation column based on duns number and row number
def segmentation():
    list1=[]
    
    for value1,v in enumerate(df16Dec['duns_number__c']):
        
        for value2,v2 in enumerate(df2['scoreaudit_wex_fleet_ml_daily.scoreapirequest_duns']):
            
            if v == v2:
                col3_val = df2.loc[value2,'scoreaudit_wex_fleet_ml_daily.scoreapiresponse_organization_customscore_segment']
                #print(v2,col3_val)
                list1.append(col3_val)
                break
        else:
            list1.append(' ')
            #print(value1,' ')
    return list1
        
  
result1 = segmentation()

#Adding the resultant list to the desired column
df16Dec['segmentation_mapped']=result1

#adding segmentation column to 2nd position[0,1]
col_name = 'segmentation_mapped'
col_position = 1
cols = df16Dec.columns.tolist()
cols.insert(col_position, cols.pop(cols.index(col_name)))
df16Dec = df16Dec[cols]
df16Dec
```

```python
#Creating Raw score column based on duns number matching 
def raw_score_column():
    
    list2=[]
    
    for value1,v in enumerate(df16Dec['duns_number__c']):
        #row_number=1
        for value2,v2 in enumerate(df2['scoreaudit_wex_fleet_ml_daily.scoreapirequest_duns']):
            #row_number+=1
            if v == v2:
                col3_val = df2.loc[value2,'scoreaudit_wex_fleet_ml_daily.scoreapiresponse_organization_customscore_rawscore']
                #print(v,col3_val)
                list2.append(col3_val)
                break
        else:
            list2.append(' ')
            #print(value1,' ')
    return list2
        
result2 = raw_score_column()

# Adding resultant list to the desired column in the dataframe
df16Dec['scoreapiresponse_organization_customscore_rawscore']=result2


#adding segmentation column to 3rd position
col_name = 'scoreapiresponse_organization_customscore_rawscore'
col_position = 2
cols = df16Dec.columns.tolist()
cols.insert(col_position, cols.pop(cols.index(col_name)))
df16Dec = df16Dec[cols]
df16Dec
```
