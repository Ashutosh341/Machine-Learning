```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
np.random.seed(1)
df = pd.DataFrame({
    'col1': np.random.normal(0, 2, 10000),
    'col2': np.random.normal(5, 3, 10000),
    
})
```

```python
df
```

```python
from sklearn.preprocessing import  StandardScaler
```

```python
sc=StandardScaler()
```

```python
sc.fit(df)
```

```python
sc_f=sc.transform(df)
sc_f = pd.DataFrame(sc_f, columns=['col1', 'col2'])
```

```python
sc_f
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(6, 5))
ax1.set_title('Before Scaling')
sns.kdeplot(df['col1'], ax=ax1)
sns.kdeplot(df['col2'], ax=ax1)
ax2.set_title('After Standard Scaler')
sns.kdeplot(sc_f['col1'], ax=ax2)
sns.kdeplot(sc_f['col2'], ax=ax2)
```

```python
from sklearn.preprocessing import  MinMaxScaler
```

```python
m=MinMaxScaler()
```

```python
m.fit(df)
```

```python
m_f=m.transform(df)

m_f= pd.DataFrame(m_f, columns=['col1', 'col2'])
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(6, 5))
ax1.set_title('Before Scaling')
sns.kdeplot(df['col1'], ax=ax1)
sns.kdeplot(df['col2'], ax=ax1)
ax2.set_title('After Standard Scaler')
sns.kdeplot(m_f['col1'], ax=ax2)
sns.kdeplot(m_f['col2'], ax=ax2)
```

```python
from sklearn.preprocessing import  RobustScaler
r=RobustScaler()
r.fit(df)
r_f=r.transform(df)

r_f= pd.DataFrame(r_f, columns=['col1', 'col2'])
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(6, 5))
ax1.set_title('Before Scaling')
sns.kdeplot(df['col1'], ax=ax1)
sns.kdeplot(df['col2'], ax=ax1)
ax2.set_title('After Standard Scaler')
sns.kdeplot(m_f['col1'], ax=ax2)
sns.kdeplot(m_f['col2'], ax=ax2)
```

```python
from sklearn.preprocessing import  Normalizer
n=Normalizer()
n.fit(df)
n_f=n.transform(df)

n_f= pd.DataFrame(n_f, columns=['col1', 'col2'])
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(6, 5))
ax1.set_title('Before Scaling')
sns.kdeplot(df['col1'], ax=ax1)
sns.kdeplot(df['col2'], ax=ax1)
ax2.set_title('After Standard Scaler')
sns.kdeplot(m_f['col1'], ax=ax2)
sns.kdeplot(m_f['col2'], ax=ax2)
```
