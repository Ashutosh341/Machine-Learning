```python
x=['Age', 'DailyRate', 'HourlyRate', 'MonthlyIncome', 'MonthlyRate', 'TotalWorkingYears', 'YearsAtCompany']
plt.grid=True
plt.figure(figsize=(30, 20))
for i, col in enumerate(x,1):
    plt.subplot(4, 2, i)
    df[df.loc[:,"Attrition"]==0][col].plot(kind='hist',ec='black',bins=35, color='blue', label='Attrition = No', alpha=0.6).grid()
    
    df[df.loc[:,"Attrition"]==1][col].plot(kind='hist',ec='black',bins=35, color='red', label='Attrition = YES', alpha=0.6).grid()
    plt.legend()
    plt.xlabel(col)
```

```python
col = df.corr().nlargest(20, "Attrition").Attrition.index
plt.figure(figsize=(20,20))

sns.heatmap(df[col].corr(), annot=True, cmap="RdYlGn", annot_kws={"size":10})
```

```python
#points

(1)Maximum Employee who left were from R&D Department
(2)Average Age of the people who left-27 to 34
(3)Most of the left employee were male
(4)Single People are most likely to Quit
(5)Employee with JobRole- Sales Representative,Research Scientist,Sales Executive and Laboratory Technician are more likely to Quit
(6)Overtime
(7)Low Performance Rating
(8)Employee with low working years
(9)Employee with less years with current company
```

**Conclusions:**

***

- `Department` : The Employee in `Research & Development` are more likely to quit then the workers on other department.
- `Age` : The Employee with Age groupy `27-34` are more likely to quit.

- `Gender` : The `Male` are more likely to quit.

- `JobRole` : The Employee in `Research Scientist`,`Laboratory Technician`, `Sales Representative`, and `Human Resources` are more likely to quit

- `MaritalStatus` : The Employee who have `Single` marital status
- `OverTime` : The Employee who work more hours are likely to quit then others.

- `BusinessTravel` : The workers who travel Rarely are more likely to quit then other employees.


- `Experience` : The Employee with less performance rating,low working years,less years with company are more likely to quit
